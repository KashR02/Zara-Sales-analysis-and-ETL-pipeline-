Cleaning
Observations - The dataset contains minimal missing values. With 1 row missing in the Name column and 2 rows missing in the description column. 
These attributes are important for NLP models, search systems, or recommendation engines.
For numerical analysis, such as revenue or sales calculations, these missing values have a negligible impact since the relevant numeric fields are complete.
Here, 3 rows are dropped to maintain consistency.

```python
df = df.dropna(subset=['name', 'description']) 
```

```python
df.shape
```

```python
df.isnull().sum()
```

#Feature engineering # revenue exploration 
Since direct revenue data was not available, revenue was approximated as:
Revenue = Price × Sales Volume

```python
#add revenue column
df["revenue"] = df["price"]*df["Sales Volume"]
```

```python
df[['price', 'Sales Volume', 'revenue']].head()
```

```python
df['revenue'].describe() 
```

```python
#which product category produces most revenue?
df.sort_values(by='revenue', ascending=False)[['Product Category', 'revenue']].head(10)
```

```python
df.groupby('Product Category')['revenue'].sum().sort_values(ascending=False)
```

```python
df.groupby('Product Category')['revenue'].sum()
```

```python
df['Product Category'].unique() #only 1 category
```

Adjusted Approach:
Instead of category-level analysis, we focus on:
- Product positioning
- Promotion strategy
-  Price segmentation

```python
#Do promoted products generate more revenue than non-promoted products? yes
df.groupby('Promotion')['revenue'].agg(['sum','mean','median','count'])
```

Insight:
Promoted products generate slightly higher average revenue (~12–13% increase)
However, the difference is moderate, not drastic
Promotion contributes to revenue, but is not the primary driver.


```python
df.groupby('Product Position')['revenue'].agg(['sum','mean','median', 'count'])
```

```python
#relation of promotion and position and revenue 
df.groupby(['Product Position', 'Promotion'])['revenue'].agg(['sum','mean','median', 'count'])
```

```python
pd.pivot_table(
    df,
    values='revenue',
    index='Product Position',
    columns='Promotion',
    aggfunc='median'
)
```

Insight:
Promotion impact varies significantly by product position.

- In End-cap placements, promotion significantly boosts revenue
- In Front-of-Store placements, promotion has minimal additional impact

Interpretation:
High-visibility placements already drive strong performance, reducing the marginal impact of promotions.

Business Strategy:
- Prioritise promotions in low-visibility areas (e.g., End-cap)
- Avoid wasting promotions on already high-performing placements

Median revenue confirms that promotion effectiveness is not uniform and depends on product placement context.

```python
import matplotlib.pyplot as plt 
```

```python
plt.scatter(df['price'], df['Sales Volume'])
plt.xlabel("Price")
plt.ylabel("Sales Volume")
```

```python


plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.boxplot(df['price'])
plt.title("Price Distribution")

plt.subplot(1,2,2)
plt.boxplot(df['Sales Volume'])
plt.title("Sales Volume Distribution")

plt.show()
```

Insight:
There is a very weak negative correlation between price and sales volume.

Interpretation:
Price alone does not strongly influence demand.

This suggests:
Customers are not highly price-sensitive
Other factors (promotion, placement) play a bigger role

Business Strategy:
 Avoid over-optimizing pricing
Focus more on visibility and positioning strategies

```python
df['price_segment'] = pd.cut(df['price'] , bins=[0,50,100,200,500], labels = ["budget", "mid","preminum", "luxury"])
```

Insight:
The majority of products fall into the mid-price segment, indicating a strong focus on affordable fashion.

Luxury products are extremely limited.

Interpretation:
Zara positions itself as a mid-market brand balancing affordability and style.

Business Strategy:
- Continue focusing on mid-range pricing as core segment
- Use premium/luxury segments selectively for brand positioning
- Use budget products for volume generation

```python
df['price_segment'].value_counts()
```

```python
df.groupby('price_segment')['Sales Volume'].median()
```

Insight:
Sales volume declines as price increases, but mid-range products maintain the highest demand
Interpretation:
Customers are comfortable spending in the mid price range, showing low price sensitivity in this band

```python
df.groupby('price_segment')['revenue'].median()
```

Insight:
Higher price segments generate significantly higher revenue per product despite lower sales volume.
Interpretation:
Revenue is driven more by pricing power than sales volume.

```python
segment_revenue = df.groupby('price_segment')['revenue'].sum()
segment_revenue.plot(kind='bar')
plt.title("Total Revenue by Price Segment")
plt.ylabel("Total Revenue")
plt.xlabel("Price Segment")
plt.show()
```

```python
df.groupby('price_segment').agg({
    'price':'median',
    'Sales Volume':'median',
    'revenue':'median'
})
```

Analysis:
Budget → High volume, low revenue
Mid → High volume, balanced revenue
Premium → Moderate volume, high revenue
Luxury → Low volume, very high revenue

Insight:
Zara operates across multiple pricing strategies simultaneously:
Volume (budget)
Balanced (mid)
Margin-driven (premium/luxury)

```python
df.groupby('price_segment')['revenue'].median().plot(kind='bar')
plt.title("Median Revenue by Price Segment")
plt.ylabel("Revenue")
plt.xlabel("Price Segment")
plt.show()
```
